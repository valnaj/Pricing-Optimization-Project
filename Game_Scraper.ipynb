## Scraping the Website for Data

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Initialize the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://randhawa.us/games/retailer/nyu.html')
```

```python
import random 

opt_list = ['maintain','10','20','40']
maintain_radio_button = driver.find_element(By.ID, 'maintainButton')
radio_button_10 = driver.find_element(By.ID, 'tenButton')
radio_button_20 = driver.find_element(By.ID, 'twentyButton')
radio_button_40 = driver.find_element(By.ID, 'fortyButton')

results_dict = {} 

for run in range(1,2501):
    # Start the game
    driver.find_element(By.ID, 'practiceButton').click()
    options = opt_list.copy()
    for round in range(1,16):
        chosen = random.choice(options)
            # Submit the decisions
        if chosen == 'maintain':
            # Select the maintain radio button
            maintain_radio_button.click()
        elif chosen == '10':
            # Select the 10 radio button
            radio_button_10.click()
            options.remove('10')
        elif chosen == '20':
            # Select the 20 radio button
            radio_button_20.click()
            options.remove('20')
        elif chosen == '40':
            # Select the 40 radio button
            radio_button_40.click()
            options.remove('40')
            break

    #get the results table and save in pandas dataframe
    results_table = driver.find_element(By.ID, 'result-table')
    results_html = results_table.get_attribute('outerHTML')
    results_df = pd.read_html(results_html, header=0)[0]
    results_df['Run_Number'] = run
    results_dict[run] = {"runs":results_df}

    #get results from the full run
    achieved_revenue = driver.find_element(By.ID, 'rev').text.strip().replace('$', '').replace(',', '')
    optimal_revenue = driver.find_element(By.ID, 'perfect').text.strip().replace('$', '').replace(',', '')
    difference = driver.find_element(By.ID, 'percentage').text.strip().replace('Difference: ', '').replace('%', '').replace(',', '')
    data = {
    'Achieved_Revenue': [float(achieved_revenue)],
    'Optimal_Revenue': [float(optimal_revenue)],
    'Difference': [float(difference)]
    }

    # Create a DataFrame
    df = pd.DataFrame(data)
    df['Run_Number'] = run
    results_dict[run]["Run_Details"] = df
    time.sleep(0.3)
    print(f"Run {run} completed")
```

```python
full_runs = results_dict[1]["runs"].copy()
all_details = results_dict[1]["Run_Details"].copy()
for key in results_dict:
    if key == 1:
        continue
    else:
        full_runs = pd.concat([full_runs, results_dict[key]["runs"]])
        all_details = pd.concat([all_details, results_dict[key]["Run_Details"]])
```

```python
full_runs.to_csv('full_runs.csv', index=False)
all_details.to_csv('all_details.csv', index=False)
```

## Calculating Demand Metrics

```python
import pandas as pd
import math
import numpy as np

df = pd.read_csv('full_runs.csv')
```

```python
#filter out rows with 0 inventory
df_filtered = df[df['Remaining Inventory'] != 0]
df_filtered = df_filtered[df_filtered['Run_Number'] <= 2500]

#Find the average Percentage demand jump after every price change between the price 60 and a discounted price
price60_avg = math.ceil(df_filtered[df_filtered['Price'] == 60]['Sales'].mean()) #average weekly sales for price 60
price54_avg = math.ceil(df_filtered[df_filtered['Price'] == 54]['Sales'].mean()) #average weekly sales for price 54
price48_avg = math.ceil(df_filtered[df_filtered['Price'] == 48]['Sales'].mean()) #average weekly sales for price 48
price36_avg = math.ceil(df_filtered[df_filtered['Price'] == 36]['Sales'].mean()) #average weekly sales for price 36


print("Average Percetage Demand Lift from price 60 to 54 (10% discount): ", math.ceil(((price54_avg - price60_avg)/price60_avg)*100))
print("Average Percetage Demand Lift from price 60 to 48 (20% discount): ", math.ceil(((price48_avg - price60_avg)/price60_avg)*100))
print("Average Percetage Demand Lift from price 60 to 36 (40% discount): ", math.ceil(((price36_avg - price60_avg)/price60_avg)*100))
print("Average Weekly Sales for Price 60: ", price60_avg)
print("Standard Deviation of Weekly Sales for Price 60: ", np.round(df_filtered[df_filtered['Price'] == 60]['Sales'].std()))
```
